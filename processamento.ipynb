```python

import pandas as pd
import numpy as np
```

```python

FILE_PATH = '/content/drive/MyDrive/TCC/Dados dissertação FINALIZADOS 24.10 (1).xlsx'
PREPROCESSED_FILE = "'/content/drive/MyDrive/TCC/dados_leite_preprocessados.csv"
ESSENTIAL_COLUMNS = ["Proteina", "Lactose", "Gordura", "CCS"]
```

```python
def transform_ccs(ccs_value):
    if pd.isna(ccs_value):
        return np.nan
    if ccs_value >= 1000:
        return 1_000_000
    else:
        return ccs_value * 1000
```

```python
dados = pd.read_excel(FILE_PATH)
print(dados.head())
dados.info()

# 1. Transformação da Coluna CCS
if 'CCS' in dados.columns:
    dados['CCS'] = dados['CCS'].apply(transform_ccs)
    print(dados[['CCS']].head(10))
else:
    print("Erro.")

linhas_antes = dados.shape[0]
dados.dropna(subset=ESSENTIAL_COLUMNS, inplace=True)
linhas_depois = dados.shape[0]

#
print(dados.head())
dados.info()
print(dados[ESSENTIAL_COLUMNS].isnull().sum())


OUTPUT_FILE = "/content/drive/MyDrive/TCC/dados_leite_preprocessados.csv"

dados.to_csv(OUTPUT_FILE, index=False)
```
